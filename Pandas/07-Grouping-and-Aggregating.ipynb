## Groupby

* if you want specific results based on some specific columns then we need to group on that specific column using groupby 
function
* It involves splitting the object, applying the function and combining the results

```python
df['Country'].value_counts()
```

```python
#this object contains a bunch of groups
df.groupby(['Country'])
```

```python
#create a variable of group of countries
country_group = df.groupby(['Country'])
```

```python
#group by country name United States
country_group.get_group('United States')
```

```python
#group by country name india
country_group.get_group('India')
```

```python
#using another way to group 
filt = df['Country'] == 'India'
df.loc[filt]
```

### Percentage of people from each country who use Python

```python
#total number of respondents from each country
country_respondents = df['Country'].value_counts()
country_respondents
```

```python
#total number of people from each country who knows python
country_uses_python = country_group['LanguageWorkedWith'].apply(lambda x: x.str.contains('Python').sum())
country_uses_python
```

```python
#create dataframe of count of people from that country and count of people who worked with python
python_df = pd.concat([country_respondents, country_uses_python], axis='columns', sort=False)
python_df
```

Here, 12469 counts of people are from united states and out of that 5964 counts of people works with python

```python
#update meaningful name of columns of python_df
python_df.rename(columns = {'Country': 'NumRespondents', 'LanguageWorkedWith': 'NumKnowsPython'}, inplace=True)
python_df
```

```python
#create a python percentage column  
python_df['PercentKnowsPython'] = (python_df['NumKnowsPython']/python_df['NumRespondents']) * 100
```

```python
python_df
```

```python
#sort countries by largest percentage of respondent who knows python
python_df.sort_values(by='PercentKnowsPython', ascending=False, inplace=True)
python_df
```
