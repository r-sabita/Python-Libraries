```python
students = {
    "firstName": ['sabita', 'bhawana', 'puja', 'sangita'],
    "lastName": ['rajbanshi', 'singh', 'oshin', 'magar'],
    "email": ['sabita@gmail.com', 'bhawana@email.com', 'puja@email.com', 'sangita@yahoo.co.uk']
    
} 
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(students)
```

```python
df
```

```python
#add firstname and lastname as fullname
df['fullName'] = df['firstName'] + ' ' + df['lastName']
```

```python
df
```

```python
#delete multiple columns of dataframe
df.drop(columns=['firstName', 'lastName'], inplace=True)
df
```

```python
#split fullname
df['fullName'].str.split(' ', expand=True)
```

```python
#add column names
df[['first','last']] = df['fullName'].str.split(' ', expand=True)
df
```

```python
#add single row of data using append
df.append({'fullName': 'aru rajbanshi'}, ignore_index=True)
```

```python
#creating another dataframe
students = {
    'first': ['pihu', 'ellina'],
    'last': ['wiz', 'stark'],
    'email': ['pihu@gmail.com', 'ellina@gmail.com']
}
df2 = pd.DataFrame(students)
```
