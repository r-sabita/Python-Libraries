# Analysis of json data

```python
students = {
    "firstName": ['sabita', 'bhawana', 'puja', 'pooza', 'pooja'],
    "lastName": ['rajbanshi', 'singh', 'magar', 'magar', 'magar'],
    "email": ['sabita@gmail.com', 'bhawana@email.com', 'puja@email.com', 'sangita@yahoo.co.uk','puja@gmail.com']
    
} 
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(students)
df
```

```python
#sort this data by firstName (default=ascending order)
df.sort_values(by='firstName')
```

```python
#sort in descending order
df.sort_values(by='firstName', ascending=False)
```

```python
df.sort_values(by=['lastName', 'firstName'], ascending=False)
```

```python
#sort lastName by descending and firstName by ascending order
#first lastname is in descending order when values in lastname are same then decide from firstname in ascending order
df.sort_values(by=['lastName', 'firstName'], ascending=[False, True])
```

```python
#sort this permananently
df.sort_values(by=['lastName', 'firstName'], ascending=[False, True], inplace=True)
```

```python
df
```

```python
#sort the index of df
df.sort_index()
```
